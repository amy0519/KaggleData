# 麥當勞早餐資料

```python
import pandas as pd
def  load_csv_data(file):
    return pd.read_csv(file)

menu = load_csv_data("menu.csv")

#head列出頂端列(資料類別)和前幾行資料
menu.head()
```

```python
#info顯示datatype、number、null值
menu.info() 
```

```python
#做個簡單的統計
menu["Category"].value_counts()
```

```python
#利用matplotlib繪圖
%matplotlib inline
import matplotlib.pyplot as plt

menu.hist(bins=100,figsize=(25,20))
plt.show()
```

### 觀察糖分與卡路里的相依關係

```python
menu.plot(kind="scatter", x="Sugars", y="Calories",alpha=0.5)
#alpha 為點組透明度，重複高-->顏色深
```

### 觀察卡路里與肥胖的相依關係

```python
menu.plot(kind="scatter", x="Calories", y="Total Fat",alpha=0.5)
```

```python
menu.plot(kind="scatter", x="Total Fat", y="Calories", alpha=0.6, 
         s=menu["Trans Fat"]/menu["Total Fat"]*1000, label="Trans Fat rate",
         c=menu["Saturated Fat"]/menu["Total Fat"]*1000, cmap=plt.get_cmap("jet"),
         colorbar = True)

axes = plt.gca()
axes.set_xlim([0,100])
axes.set_ylim([100,1000])

plt.show()
```
